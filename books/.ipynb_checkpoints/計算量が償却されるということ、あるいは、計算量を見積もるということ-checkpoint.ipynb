## ä¾‹ 1

$$
\begin{aligned}
f(n) & = 3n^2 + n + 100 \\
g(n) & = n^2
\end{aligned}
$$

ã¨ã™ã‚‹ã¨ã€

$$
\begin{aligned}
\lim_{n \rightarrow \infty}\frac{f(n)}{g(n)}ã€€= lim_{n \rightarrow \infty}\frac{3n^2 + n + 100}{n^2} = lim_{n \rightarrow \infty}{3 + \frac{1}{n} + 100 \frac{1}{n^2}} = 3
\end{aligned}
$$

ã¨ãªã‚‹ã€‚

ã“ã‚Œã¯æœ‰ç•Œãªã®ã§ $f=O(g)=O(n^2)$ ã¨è©•ä¾¡ã•ã‚Œã‚‹ã€‚

```python
!pip install matplotlib numpy
```

```python
import matplotlib.pyplot as plt
import numpy as np

n = np.arange(0, 10, 0.1)

plt.plot(n, 3 * n * n + n + 100, label=r"$3n^2 + n + 100$")
plt.plot(n, n * n, label=r"$n^2$")
plt.plot(n, n, label=r"$n$")
plt.legend(bbox_to_anchor=(0, 1), loc='upper left', borderaxespad=1)
plt.show()
```

## ã§ã€ã©ã†ä½¿ã†ã®ï¼Ÿ

```python
# O(n) ã¨è©•ä¾¡ã•ã‚Œã‚‹ã‚³ãƒ¼ãƒ‰

# len(arr) = n
def find(arr, x):
    for i, y in enumerate(arr): # n å›žãƒ«ãƒ¼ãƒ—ãŒå›žã‚‹
        if x == y:
            return i # ãªã‚“ã‹è¨ˆç®—ã—ã¦ã‚‹ã£ã½ã„ã‘ã©ã€æ‰€è©®å®šæ•°å€ã ã¨æ€ã£ã¦ "1" ã ã¨æ€ã†
```

```python
# O(log n) ã¨è©•ä¾¡ã•ã‚Œã‚‹ã‚³ãƒ¼ãƒ‰

# len( (ng,ok] ) = n
def binary_search(is_ok, ng, ok):
    while (abs(ok - ng) > 1):
        mid = (ok + ng) // 2 # ng ã¨ ok ã®ä¸­å¤®ã®å€¤ã«ã¤ã„ã¦èª¿æŸ»
        if is_ok(mid):
            ok = mid
        else:
            ng = mid # ã©ã¡ã‚‰ã®å ´åˆã§ã‚‚æŽ¢ç´¢ç¯„å›²ãŒåŠåˆ†ã«ãªã‚‹
    return ok
```

```python
# https://atcoder.jp/contests/abc051/tasks/abc051_b
# 0~k ã®æ•°å­—ã®å†…ã§ x+y+z = s ã¨ãªã‚‹ã‚‚ã®ã®æ•°ã‚’æ±‚ã‚ã‚ˆ

# O(k^3)
def solve(k, s):
    ans = 0
    for x in range(k+1):
        for y in range(k+1):
            for z in range(k+1):
                if x+y+z == s:
                    ans += 1
    return ans
```

```python
# https://atcoder.jp/contests/abc051/tasks/abc051_b
# 0~k ã®æ•°å­—ã®å†…ã§ x+y+z = s ã¨ãªã‚‹ã‚‚ã®ã®æ•°ã‚’æ±‚ã‚ã‚ˆ

# O(k^2)
def solve(k, s):
    ans = 0
    for x in range(k+1):
        for y in range(k+1):
            z = s - x - y
            if 0 <= z and z <= k:
                ans += 1
    return ans
```

```python
import matplotlib.pyplot as plt
import numpy as np

n = np.arange(0, 1000, 1)

plt.plot(n, n * n * n, label=r"$n^3$")
plt.plot(n, n * n, label=r"$n^2$")
plt.legend(bbox_to_anchor=(0, 1), loc='upper left', borderaxespad=1)
plt.show()
```
